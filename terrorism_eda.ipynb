```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r"terrorism_data.csv", encoding='latin1')
```

```python
df.head()
```

```python
df.describe()
```

```python
#displaying shape and column names
print(df.shape)
print(df.columns.tolist())
```

```python
#displaying number of unique categories in each column
df.nunique()
```

```python
df1=df
```

```python
#removing all columns with more than 1000 null values
for i in df.columns:
    if(df[i].isnull().sum()>1000):
        df1.drop(i, axis=1, inplace=True)
```

```python
df1.head()
```

```python
print(df1.shape)
print(df1.columns.tolist())
```

```python
#checking null value counts
for i in df1.columns:
    print(i, df[i].isnull().sum())  
```

```python
#removing eventid and iday as they are not very relevant
df1=df1.drop(['eventid', 'iday'], axis=1)
```

```python
#correlation map of all variables; the darker portions indicate higher correlation
cmap=sns.diverging_palette(1220, 510, as_cmap=True)
sns.heatmap(df1.corr(), vmax=0.3, cmap=cmap, center=0, square=True, linewidths=0.5, cbar_kws={"shrink":0.5})
```

```python
#correlation matrix
df1.corr()
```

```python
#plotting bargraphs for all data columns
for i in df1.columns:
    if(len(df1[i].unique())>=20):
        cat_num=df[i].value_counts()[:20]
        print("graph for %s total: %d" %(i, len(cat_num)))
        chart=sns.barplot(x=cat_num.index, y=cat_num)
        chart.set_xticklabels(chart.get_xticklabels(), rotation=90)
        plt.show()
    else:
        cat_num=df1[i].value_counts()
        print("graph for %s total: %d" %(i, len(cat_num)))
        chart=sns.barplot(x=cat_num.index, y=cat_num)
        chart.set_xticklabels(chart.get_xticklabels(), rotation=90)
        plt.show()
```
